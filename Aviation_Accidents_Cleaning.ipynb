## Constructing Key Injury Metric

**Approach:**  
We estimate total occupants as the sum of all four injury columns (Fatal + Serious + Minor + Uninjured). This gives a per-flight denominator even when seat capacity is not recorded.

**Imputation assumption:** NaN values in injury columns are treated as 0. An unreported injury category most likely means zero occurrences in that category (NTSB reports are comprehensive when injuries exist).

**Derived column — `Serious_Fatal_Frac`:** Fraction of occupants who were fatally or seriously injured.  
Formula: `(Fatal + Serious) / Total_Aboard`. Clipped to [0, 1] to guard against any data entry errors.

```python
injury_cols = ['Total.Fatal.Injuries','Total.Serious.Injuries',
               'Total.Minor.Injuries','Total.Uninjured']

for col in injury_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

df['Total.Aboard'] = df[injury_cols].sum(axis=1)

# Remove rows with no recorded occupants (uninformative for injury rate)
df = df[df['Total.Aboard'] > 0].copy()

df['Serious_Fatal_Frac'] = (
    (df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries'])
    / df['Total.Aboard']
).clip(0, 1)

print('Serious_Fatal_Frac summary:')
df['Serious_Fatal_Frac'].describe()
```
